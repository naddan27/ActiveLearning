```python
import pandas as pd
import numpy as np
import os
from matplotlib import pyplot as plt
from glob import glob
```

```python
basedir = "C:/Users/danny/Documents/Bai/activelearning/metrics"
testing = "uncertainty_bootstrapping"

folders = glob(os.path.join(basedir, testing, "iteration_*"))
plt.figure(figsize = (3, 8))
plt.suptitle("Variance of Bootstrapped Predictions")
kwargs = dict(alpha=1, bins=10, density=False, stacked=False)

for i, x in enumerate(folders):
    csv_path = glob(os.path.join(x, "uncertainty_*"))[0]
    print("Reading csv from", csv_path)

    uncertainty_df = pd.read_csv(csv_path)
    
    all_variances = uncertainty_df["Variance"]
    selected_variances = uncertainty_df[uncertainty_df["Selected"] == "Y"]["Variance"]
    nonselected_variance = uncertainty_df[uncertainty_df["Selected"] == "N"]["Variance"]
    
    all_variances = np.sqrt(all_variances)
    selected_variances = np.sqrt(selected_variances)
    nonselected_variance = np.sqrt(nonselected_variance)
    
    plt.subplot(len(folders), 1, i+1)
    plt.title("Iteration " + os.path.basename(x).split("_")[1])
    plt.hist(nonselected_variance, **kwargs, label = "Not Selected")
    plt.hist(selected_variances, **kwargs, label = "Selected")
    plt.xlim((0,0.1))
plt.tight_layout()
```

```python
basedir = "C:/Users/danny/Documents/Bai/activelearning/metrics"
testing = "uncertainty_prob_roi"

folders = glob(os.path.join(basedir, testing, "iteration_*"))
plt.figure(figsize = (3, 8))
plt.suptitle("Probability at ROI")
kwargs = dict(alpha=1, density=False, stacked=True)

for i, x in enumerate(folders):
    csv_path = glob(os.path.join(x, "uncertainty_*"))[0]
    print("Reading csv from", csv_path)

    uncertainty_df = pd.read_csv(csv_path)
    
    all_variances = uncertainty_df["Prob_at_ROI"]
    selected_variances = uncertainty_df[uncertainty_df["Selected"] == "Y"]["Prob_at_ROI"]
    nonselected_variance = uncertainty_df[uncertainty_df["Selected"] == "N"]["Prob_at_ROI"]
    selected_variances[selected_variances < 0.00001] = 0
    nonselected_variance[nonselected_variance < 0.00001] = 0
    
    plt.subplot(len(folders), 1, i+1)
    plt.title("Iteration " + os.path.basename(x).split("_")[1])
    plt.hist(nonselected_variance, **kwargs, label = "Not Selected")
    plt.hist(selected_variances, **kwargs, label = "Selected")
plt.tight_layout()
```
